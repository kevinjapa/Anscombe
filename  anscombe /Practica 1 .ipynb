```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cargar el dataset
file_path = 'df_anscombe.csv'
df = pd.read_csv(file_path)
print("Dataset Cargado")
```

```python
df.groupby('group').describe()
```

## Graficar los 4 graficos de cada categoria y los 4 blockpot

### Graficos de Cada Categoria

```python
sns.set(style="ticks")

fig, axes = plt.subplots(1, 4, figsize=(16, 5))

for i, group in enumerate(df['group'].unique()):
    group_data = df[df['group'] == group]
    sns.scatterplot(x='x', y='y', data=group_data, ax=axes[i], color='blue')
    sns.lineplot(x=[group_data['x'].quantile(0.25), group_data['x'].quantile(0.25)],
                 y=[group_data['y'].min(), group_data['y'].max()], ax=axes[i], color='red', linestyle='--', label="Q1")
    sns.lineplot(x=[group_data['x'].quantile(0.50), group_data['x'].quantile(0.50)],
                 y=[group_data['y'].min(), group_data['y'].max()], ax=axes[i], color='green', linestyle='--', label="Q2")
    sns.lineplot(x=[group_data['x'].quantile(0.75), group_data['x'].quantile(0.75)],
                 y=[group_data['y'].min(), group_data['y'].max()], ax=axes[i], color='orange', linestyle='--', label="Q3")

    axes[i].set_title(f"Grupo {group}")
    axes[i].legend()

plt.tight_layout()
plt.show()
```

### Grafico de Cajas y Bigotes

```python
sns.set(style="ticks")

fig, axes = plt.subplots(1, 4, figsize=(16, 5))

for i, group in enumerate(df['group'].unique()):
    group_data = df[df['group'] == group]
    sns.boxplot(y='y', data=group_data, ax=axes[i])
    axes[i].set_title(f"Caja y Bigotes Grupo: {group}")
plt.tight_layout()
plt.show()
```
